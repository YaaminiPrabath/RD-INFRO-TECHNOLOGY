```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel("Real_Estate_Sales_dataset.xlsx")
data=pd.DataFrame(df)
yearly_mean=data.groupby('List Year')['Sale Amount'].mean()
plt.plot(yearly_mean.index,yearly_mean.values,color='blue')
plt.title("Average Sales amount by year")
plt.xlabel("Year")
plt.ylabel("Average sales amount")
plt.show()
```

```python
hist_plot=data['Sales Ratio']
plt.hist(hist_plot,bins=30,color='pink',edgecolor='black')
plt.title('Distribution of Sales  Ratio')
plt.xlabel("Sale Ratios")
plt.ylabel('Number of properties')
plt.show()
```

```python
import seaborn as sns
data1=data[['Assessed Value','Sale Amount']].corr()
sns.heatmap(data1,cmap='viridis',annot=True)
plt.title("correlation between assessed value and sales amount")
plt.show()
```
